#### Display the trending Python repositories in GitHub

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url1 = 'https://github.com/trending/python?since=daily'
````

````python
#your code
dev_py = requests.get(url1)
dev_py.content
soup_dev_py = BeautifulSoup(dev_py.content)
(soup_dev_py.prettify())
````

````python
repos_names = soup_dev_py.select('span.text-normal')
results = [name.text.strip('\n ').replace('/' ,'') for name in repos_names]
results
````

#### Display all the image links from Walt Disney wikipedia page

````python
# This is the url you will scrape in this exercise
url2 = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
#your code
walt_req = Request(url2)
html_page = urlopen(walt_req)

soup_walt =BeautifulSoup(html_page, "lxml")
(soup_walt.prettify())

walt_links = []
for link in soup_walt.findAll('a', class_="image"):
    walt_links.append(link.get('href'))
    
walt_links
````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
url3 ='https://en.wikipedia.org/wiki/Python' 
````

````python
#your code
pyth_req = Request(url3)
html_page = urlopen(pyth_req)

soup_pyth =BeautifulSoup(html_page, "lxml")
(soup_pyth.prettify())

links_list = []
for link in soup_pyth.select('li > a'):
    links_list.append(link.get('href').strip('\n '))
    
links_list
````

#### A Python list with the top ten FBI's Most Wanted names

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url6 = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
lat_eart = requests.get(url6).content
earth_soup = BeautifulSoup(lat_eart)
earth_soup.select('#tbody')
table = pd.read_html(lat_eart)[3]
my_table = table.head(20)[['Date & Time UTC', 'Latitude degrees','Longitude degrees','Last update [-]']]
my_table.columns = [['Date & Time UTC', 'Lat deg',' Lat Dir', 'Long deg','Long Dir','Last update [-]']]

my_table
````

#### Display the date, days, title, city, country of next 25 hackathon events as a Pandas dataframe table

````python
# This is the url you will scrape in this exercise
url ='https://mlh.io/seasons/na-2020/events'
````

````python

reqs = requests.get(url)
hack_soup = BeautifulSoup(reqs.content,'lxml')
(hack_soup.prettify())
hack_info =hack_soup.find_all('div', {'class':'event-wrapper'})

for x, y in enumerate (hack_info [0:25]):
    date_event = y.find('p',{'class':'event-date'}).text.strip()
    title = y.find('h3', {'itemprop':'name'}).text.strip()
    city = y.find('span', {'itemprop':'city'}).text.strip()
    country = y.find('span',{'itemprop':'state'}).text.strip()
    test =("{}{}, {}, {}, {} \n". format(str(x)+ ')', date_event, title, city, country))
    print(test)
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
html = requests.get(url).content
soup = BeautifulSoup(html,"lxml")
````

````python
#your code
languages =soup.select('a.link-box > strong')
lang1 = [lang1.text for lang1 in languages]
articles = soup.select ('a.link-box > small')
arti1 = [re.sub(r'\xa0|\+', '', num_art.bdi.text) for num_art in articles]
list(zip(lang1, arti1))
````
